21\. İçerisinde \[AKTIFSPOR\] dan alışveriş yapmış kişilerin Order Channel’larını sağ tarafa kolon olarak ekleyen sorguyu <span style="color: var(--vscode-foreground);">yazınız.</span>

```sql
SELECT *
FROM (
    SELECT [master_id], [order_channel]
    FROM [dbo].[FLO]
    WHERE [interested_in_categories_12] LIKE '%AKTIFSPOR%'
) AS SourceTable
PIVOT (
    COUNT([order_channel])
    FOR [order_channel] IN ([Android App], [Ios App], [Mobile], [Desktop])
) AS PivotTable;
```
